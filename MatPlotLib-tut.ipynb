### Load Necessary Libraries

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

### Basic Graphs

```python
# import gas price data

gas_df = pd.read_csv("C:/Users/GREEN LION/Documents/Python Tutorials/Keith Galli - Data Science/gas_prices.csv")
```

```python
gas_df
```

```python
# replace the missing price in Austria with the mean price

gas_df.Australia.fillna(gas_df.Australia.mean(),inplace=True)
```

```python
gas_df
```

### Line Graph

```python
#ploting the gas prices over time in the below countries

plt.figure(figsize=(8,5))
plt.plot(gas_df['Year'],gas_df['USA'], label='USA')
plt.plot(gas_df['Year'],gas_df['Canada'],label='Canada')
plt.plot(gas_df['Year'],gas_df['Italy'], label='Italy')

plt.xticks(gas_df.Year[::3])
plt.title('Gas Price over Time from 1990-2008 in USA, Canada and Italy.')
plt.xlabel('Year')
plt.ylabel('Gas Price (in USD)')


plt.legend()

plt.show()
```

```python
# compare other countries

plt.figure(figsize=(8,5))
plt.plot(gas_df['Year'],gas_df['France'], label='France',color ='r',marker='.')
plt.plot(gas_df['Year'],gas_df['Germany'],label='Germany',color ='y',marker='.')
plt.plot(gas_df['Year'],gas_df['Japan'], label='Japan',color ='black',marker='.')

plt.xticks(gas_df.Year[::3])
plt.title('Gas Price over Time from 1990-2008 in France, Germany and Japan.')
plt.xlabel('Year')
plt.ylabel('Gas Price (in USD)')


plt.legend()

plt.show()
```

### Scatter plot

```python
# checking relationship btn prices in Japan and South Korea

plt.scatter(gas_df.USA, gas_df.Japan, color ='r',alpha=1,s=50, marker ='*')

plt.title('Relationship between Prices in Japan and USA')
plt.xlabel('Japan')
plt.ylabel('USA')

plt.show()
```

### Bar Chart

```python
# Unpivot the data from several columns into two(rows, keeping the year.
new_df = pd.melt(gas_df, 
                 id_vars=['Year'], 
                 var_name='Country', 
                 value_name='Price')
```

```python
new_df.head()
```

```python
# grouping the country by price
sorted_df = new_df.groupby('Country', as_index=False)['Price'].sum()     #.reset_index()
sorted_df = sorted_df.sort_values(by='Price', ascending=True)

plt.figure(figsize=(8,5))
plt.barh(sorted_df['Country'], sorted_df['Price'])    
plt.xlabel('Gas Prices ')
plt.title('Total Price by Country')

plt.show()
```

 # Loading the Fifa Data

```python
fifa_df = pd.read_csv("C:/Users/GREEN LION/Documents/Python Tutorials/Keith Galli - Data Science/fifa_data.csv")
```

```python
pd.set_option('display.max_columns',89)

fifa_df.head(30)
```

```python
# removing the Unnamed: 0 column

fifa_df.drop(columns = 'Unnamed: 0',inplace=True)
```

### Histogram

```python
# plotint the overall skill levels of players
plt.figure(figsize=(8,5))

bins = range(40,110,10)
plt.hist(fifa_df['Overall'],bins= bins, color = 'b')


plt.title('Distribution of Players in Fifa 2018')
plt.xlabel('Skill Level(Overall)')
plt.ylabel('Number of players')

plt.show()
```

### Pie Chart

```python
#ploting the preferred foot of players

No_left = fifa_df.loc[fifa_df['Preferred Foot']=='Left'].count()[0]
No_right = fifa_df.loc[fifa_df['Preferred Foot']=='Right'].count()[0]
none = fifa_df.loc[fifa_df['Preferred Foot'].isna()].count()[0]
print(No_left)
print(No_right)
print(none)
print(No_right+No_left)
```

```python
plt.pie([No_left,No_right,none], labels = ['Left', 'Right','none' ], colors =['#93a832','#3281a8','r'], autopct ='%1.1f%%',\
        startangle = 50)

plt.title('Foot Preference of Players')


plt.show()
```

```python
# Categories the weight of the players into brackets
#split the lbs from the numbers

fifa_df.Weight = [int(x.strip('lbs')) if type(x)==str else x for x in fifa_df.Weight]

# group the weights into light, medium and heavy
plt.figure(figsize=(8,5))
plt.style.use('ggplot')

light = fifa_df.loc[fifa_df.Weight<150].count()[0]
medium = fifa_df.loc[(fifa_df.Weight>=150) & (fifa_df.Weight<=200)].count()[0]
heavy = fifa_df.loc[(fifa_df.Weight>200)].count()[0]

weight = [light,medium,heavy]
label = ['Below 150','150-200','Over 200']

plt.title('Weight Distribution of Players')

plt.pie(weight, labels = label, autopct = '%1.1f%%', startangle=45)

plt.show()
```

### Box and whisker Chart

```python
# Comparing teams in England with the overall score

plt.style.use('default')
plt.figure(figsize=(8,5))
Chelsea = fifa_df.loc[fifa_df.Club == 'Chelsea']['Overall']
Man_U = fifa_df.loc[fifa_df.Club == 'Manchester United']['Overall']
Arsenal = fifa_df.loc[fifa_df.Club == 'Arsenal']['Overall']
 
label = ['Chelsea','Man_U', 'Arsenal']
plt.boxplot([Chelsea,Man_U,Arsenal], labels = label)

plt.title('Professional Football Team Comparison')
plt.ylabel('Overall Skill Level')
plt.xlabel('Teams')
plt.show()
```
